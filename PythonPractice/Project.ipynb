```python
import math
import random
```

```python
def FirstQ(func1, func2, x, n, m, a, b, outf):
    outf.write("Вопрос 1.\nУсловие:\n")
    names1 = dict()
    names1 = {0:"{x}^{%s}"%a, 1:"e^{x}", 2:"{%s}^{x}"%a, 3:"ln({x})", 4:"log_{%s}({x})"%a, 5:"sin({x})", 6:"cos({x})", 7:"tg({x})", 8:"ctg({x})"}

    names2 = dict()
    names2 = {0:"{x}^{%s}"%b, 1:"e^{x}", 2:"{%s}^{x}"%b, 3:"ln({x})", 4:"log_{%s}({x})"%b, 5:"sin({x})", 6:"cos({x})", 7:"tg({x})", 8:"ctg({x})"}
    
    q = random.randint(0,2)
    
    funcf = dict()
    funcf = {names1[0]: n * x**(a), names1[1] : n * math.exp(x), names1[2] : n * a**(x), names1[3] : n * math.log(x), names1[4] : n * math.log(x, a), names1[5] : n * math.sin(x), names1[6] : n * math.cos(x), names1[7] : n * math.tan(x), names1[8] : n / math.tan(x)}

    df = dict()
    df = {names1[0]: n * a * x**(a-1), names1[1] : n * math.exp(x), names1[2] : n * a**(x) * math.log(a), names1[3] : n /  x, names1[4] : n / (math.log(a)*x), names1[5] : n * math.cos(x), names1[6] : (-1) *n * math.sin(x), names1[7] : n / math.cos(x)**2, names1[8] : (-1) * n / math.sin(x)**2}

    funcg = dict()
    funcg = {names2[0]: m * x**(b), names2[1] : m * math.exp(x), names2[2] : m * b**(x), names2[3] : m * math.log(x), names2[4] : m * math.log(x, b), names2[5] : m * math.sin(x), names2[6] : m * math.cos(x), names2[7] : m * math.tan(x), names2[8] : m / math.tan(x)}

    dg = dict()
    dg = {names2[0]: m * b * x**(b-1), names2[1] : m * math.exp(x), names2[2] : m * b**(x) * math.log(b), names2[3] : m /  x, names2[4] : m / (math.log(b)*x), names2[5] : m * math.cos(x), names2[6] : (-1) * m * math.sin(x), names2[7] : m / math.cos(x)**2, names2[8] : (-1) * m / math.sin(x)**2}
    
    if(q==0):
        f = names1[func1]
        g = names2[func2]
        print("Вопрос 1.\nНайдите значение производной функции $f(x) = %s * %s + (%s) * %s$ в точке $x_0$ = %s. В ответ запишите только целую часть." %(n,f,m,g,x))
        outf.write("Найдите значение производной функции $f(x) = %s * %s + (%s) * %s$ в точке $x_0$ = %s. В ответ запишите только целую часть.\n" %(n,f,m,g,x))

        try:
            answer = input()
            if(int(answer) == math.floor(df[names1[func1]]+dg[names2[func2]])):
                ans1 = "Правильно"
            else:
                ans1 = "Неправильно"
        except ValueError as e:
            ans1 = "Неправильно"
            print("Недопустимое значение ответа! Ответ засчитан как неправильный")
        outf.write("Правильный ответ: %s\nОтвет пользователя: %s\n" %(math.floor(df[names1[func1]]+dg[names2[func2]]),answer))
    if(q==1):
        f = names1[func1]
        g = names2[func2]
        print("Вопрос 1.\nНайдите значение производной функции $f(x) = %s * %s * (%s) * %s$ в точке $x_0$ = %s. В ответ запишите только целую часть." %(n,f,m,g,x))
        outf.write("Найдите значение производной функции $f(x) = %s * %s * (%s) * %s$ в точке $x_0$ = %s. В ответ запишите только целую часть.\n" %(n,f,m,g,x))

        print(format(math.floor(df[names1[func1]]*funcg[names2[func2]]+dg[names2[func2]]*funcf[names1[func1]])))

        try:
            answer = input()
            if(int(answer) == math.floor(df[names1[func1]]*funcg[names2[func2]]+dg[names2[func2]]*funcf[names1[func1]])):
                ans1 = "Правильно"
            else:
                ans1 = "Неправильно"
        except ValueError as e:
            ans1 = "Неправильно"
            print("Недопустимое значение ответа! Ответ засчитан как неправильный")
        outf.write("Правильный ответ: %s\nОтвет пользователя: %s\n" %(math.floor(df[names1[func1]]*funcg[names2[func2]]+dg[names2[func2]]*funcf[names1[func1]]),answer))
    if(q==2):
        f = names1[func1]
        g = names2[func2]
        print("Вопрос 1.\nНайдите значение производной функции $f(x) = {{%s * %s} \over {%s * %s}}$ в точке $x_0$ = %s. В ответ запишите только целую часть. Если значения не сушествует введите восклицательный знак." %(n,f,m,g,x))
        outf.write("Найдите значение производной функции $f(x) = {{%s * %s} \over {%s * %s}}$ в точке $x_0$ = %s. В ответ запишите только целую часть. Если значения не сушествует введите восклицательный знак.\n" %(n,f,m,g,x))
        if(funcg[names2[func2]] != 0):
            rightans1 = math.floor((df[names1[func1]]*funcg[names2[func2]]-dg[names2[func2]]*funcf[names1[func1]])/funcg[names2[func2]]**2)
        else:
            rightans1 = "!"

        try:
            answer = input("Ответ: ")
            if(rightans1!="!"):
                if(int(answer) == rightans1):
                    ans1 = "Правильно"
                else:
                    ans1 = "Неправильно"
            else:
                if(answer == rightans1):
                    ans1 = "Правильно"
                else:
                    ans1 = "Неправильно"
        except ValueError as e:
            ans1 = "Неправильно"
            print("Недопустимое значение ответа! Ответ засчитан как неправильный")
        outf.write("Правильный ответ: %s\nОтвет пользователя: %s\n" %(rightans1,answer))
    
    return ans1
```

```python
def SecondQ(a1,b1,c1,a2,b2,c2, outf):
    outf.write("\nВопрос 2.\nУсловие:\n")
    print("Вопрос 2.\nНайдите площадь между $f(x) = %s*{x}^2 + (%s)*{x} + (%s)$ и $g(x) = %s*{x}^2 + (%s)*{x} + (%s)$. В ответ запишите только целую часть." %(a1,b1,c1,a2,b2,c2))
    outf.write("Найдите площадь между $f(x) = %s*{x}^2 + (%s)*{x} + (%s)$ и $g(x) = %s*{x}^2 + (%s)*{x} + (%s)$. В ответ запишите только целую часть.\n" %(a1,b1,c1,a2,b2,c2))
    ka = a1-a2
    kb = b1-b2
    kc = c1-c2
    D = kb**2 - 4*ka*kc
    if(D>0):
        x1 = (-kb-math.sqrt(D))/(2*ka)
        x2 = (-kb+math.sqrt(D))/(2*ka)
        rightans2=math.floor(math.fabs(((ka*x2**3)/3+(kb*x2**2)/2+kc*x2)-((ka*x1**3)/3+(kb*x1**2)/2+kc*x1)))
    else:
        rightans2 = 0
    try:
        answer = input("Ответ: ")
        if(int(answer) == rightans2):
            ans2 = "Правильно"
        else:
            ans2 = "Неправильно"
    except ValueError as e:
        ans2 = "Неправильно"
        print("Недопустимое значение ответа! Ответ засчитан как неправильный")
    outf.write("Правильный ответ: %s\nОтвет пользователя: %s\n" %(rightans2,answer))
    
    return ans2
```

```python
def ThirdQ(x,y,outf):
    outf.write("\nВопрос 3.\nУсловие:\n")
    print("Вопрос 3.\nНайдите градиент $f(x) = e^{x*y}$ в точке с координатами (%s, %s). В ответ введите сумму координат градиента." %(x,y))
    outf.write("Найдите градиент $f(x) = e^{x*y}$ в точке с координатами (%s, %s). В ответ введите сумму координат градиента.\n" %(x,y))
    rightans3 = float((y+x)*math.exp(x*y))
    try:
        answer = input("Ответ: ")
        if(int(answer) == rightans3):
            ans3 = "Правильно"
        else:
            ans3 = "Неправильно"
    except ValueError as e:
        ans3 = "Неправильно"
        print("Недопустимое значение ответа! Ответ засчитан как неправильный")
    outf.write("Правильный ответ: %s\nОтвет пользователя: %s\n\n" %(rightans3,answer))
    
    outf.close()
    return ans3
```
